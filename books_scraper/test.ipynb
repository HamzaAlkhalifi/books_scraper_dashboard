```python
from lazyscraper import Request
from selectolax.parser import HTMLParser
```

```python
req = Request(URL= "https://books.toscrape.com/", headers= None, params= None, proxies= None, cookies= None, impersonate= "chrome")
res = req.get().text

parser = HTMLParser(res)
```

```python
category = parser.css_first("h1").text()
```

```python
def parse(elements):
    for e in elements:
        if e.attributes.get('title'):
            break 
    return e.attributes.get('title')
        
for book in parser.css_first("ol.row").iter():
    list_= []
    title= parse(book.css("a"))
    price= book.css_first("p.price_color").text()
    availability= book.css_first("p.instock.availability").text()
    list_.append({
        "title": title,
        "category": category,
        "price": price,
        "availability": availability.strip()
    })
    print(list_)
    
```
